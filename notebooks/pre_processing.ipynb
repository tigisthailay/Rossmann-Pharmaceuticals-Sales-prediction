#### Handling Missing Value

fill the numeric columns  'Promo2SinceYear', 'Promo2SinceWeek', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'CompetitionDistance' with mean values

```python
#handling numeric column with median
miss = ['Promo2SinceYear', 'Promo2SinceWeek', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'CompetitionDistance']
train_df = preprocess.fill_missing_median(train_df, miss)
test_df = preprocess.fill_missing_median(test_df, miss)
```

Fill the Object column with mode

```python
#handling object column with mode
train_df = preprocess.fill_missing_mode(train_df, ['PromoInterval'])
test_df = preprocess.fill_missing_mode(test_df, ['PromoInterval'])
```

Fill Open with zero

```python
test_df = preprocess.fill_missing_with_zero(test_df, ['Open'])
```

### Label Encoding

Categorical Columns: PromoInterval, Assortment, StoreType

```python
categorical_columns = ['PromoInterval', 'Assortment', 'StoreType']
# Label encoding
label_encoded_columns = preprocess.label_encode(train_df, categorical_columns)
```

```python
train_df.drop(categorical_columns, axis=1, inplace=True)
# Merge DataFrames
train_df = pd.concat([train_df, label_encoded_columns], axis=1)
```
